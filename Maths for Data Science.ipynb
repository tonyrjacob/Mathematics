# Question Number 1: 

dx(t)/ dt = −x(t) + u(t)




u(t) = {1, t ≥ 10
        
        
        
     0, t < 10 }

# Solution:

```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Craeting a function that returns the value of dy/dt
def model_1(x,t):
    # u steps from 0 to 1 at t=10
    if t<10.0:
        u = 0
    else:
        u = 1
    dy_dt = (-x + u)
    return dy_dt

# initial condition
x0 = 1

t = np.linspace(0,40,1000)

# solve ODE
x = odeint(model_1,x0,t)

# plot results
plt.plot(t,x,'r-',label='Output (x(t))')
plt.plot([0,10,10,40],[0,0,1,1],'b-',label='Input (u(t))')
plt.ylabel('values')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```

# Question Number 2

# Question 4

# Question 5

Let’s suppose we have the following set of differential equations:



{




𝑑𝑦/𝑑𝑡 = −0.4𝑦 + 0.02𝑥𝑦
𝑑𝑥/𝑑𝑡 = 0.8𝑥 − 0.01(𝑥^2) − 0.1𝑥𝑦



}


x and y could be two populations of animals in the same area. Therefore, the growth of one 
population depends on the number of predator animals of the other species. Both initial 
populations are set to 100

# Solution:

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dz/dt
def model_5(z,t):
    x = z[0]
    y = z[1]
    dx_dt = ((0.08*x) - (0.01*(x**2)) - (0.1*x*y))
    dy_dt = ((-0.4*y) + (0.02*x*y))
    dz_dt = [dx_dt,dy_dt]
    return dz_dt

# initial condition
z0 = [100,100]

# number of time points
n = 401

# time points
t = np.linspace(0,40,n)


# store solution
x = np.empty_like(t)
y = np.empty_like(t)
# record initial conditions
x[0] = z0[0]
y[0] = z0[1]

# solve ODE
for i in range(1,n):
    # span for next time step
    tspan = [t[i-1],t[i]]
    # solve for next step
    z = odeint(model_5,z0,tspan)
    # store solution for plotting
    x[i] = z[1][0]
    y[i] = z[1][1]
    # next initial condition
    z0 = z[1]

# plot results
plt.plot(t,x,'b-',label='x(t)')
plt.plot(t,y,'r--',label='y(t)')
plt.ylabel('values')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```
